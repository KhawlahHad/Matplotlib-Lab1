```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

from datetime import datetime
import random
```

## Q1: Create two charts using list and dataframe.

```python
#DataFrame

data = {
    'Category': ['Housing', 'Transportation', 'Food', 'Entertainment', 'Utilities'],
    'Percentage': [30, 15, 20, 10, 25]
}
df = pd.DataFrame(data)
plt.figure(figsize=(4, 4)) 
plt.pie(df['Percentage'], labels=df['Category'], autopct='%1.1f%%', startangle=140)
plt.title("Household Budget Distribution")
plt.show()
```

```python
# List

accidents = [random.randint(10000, 50000) for i in range(21)]
years   = [int(i) for i in range(2000, 2021)]

plt.plot(
    years, accidents,
    marker='o', linestyle='--',
    color='r', label='Accidents Over the Years'
)

ax = plt.gca()
ax.set_xticks(years)
ax.set_xticklabels(years, rotation=45)
plt.title('Accidents Over the Years', fontsize=15)
plt.xlabel('Year', fontsize=12)
plt.ylabel('No. of accidents', fontsize=12)
plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
plt.pie(df['Percentage'], labels=df['Category'], autopct='%1.1f%%', startangle=140,  explode=[0.1] * len(df['Category']))
plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 12, 9]

plt.figure(figsize=(4, 4))  
plt.scatter(x, y, s=50, marker='o', alpha=0.4)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Scatter Chart')
plt.show()
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.figure(figsize=(4, 4))  
plt.scatter(x, y, s=50, marker='o', alpha=0.4)
plt.xlabel('X-Axis', fontsize=16)
plt.ylabel('Y-Axis', fontsize=16)
plt.title('Scatter Chart', color='red', loc='left', fontsize=16)
plt.show()
```

## Q5: Create three line charts with Legends at left and colors.

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 15, 7, 12, 9]
y2 = [5, 8, 6, 10, 7]
y3 = [12, 6, 8, 10, 15]

plt.figure(figsize=(4, 3))  
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Line Chart')
plt.plot(x, y1, label='Line 1', color='blue')
plt.plot(x, y2, label='Line 2', color='green')
plt.plot(x, y3, label='Line 3', color='red')
plt.legend(loc='upper left')

plt.show()
```

## Q6: Create subplots with three charts and figsize=20,10.

```python
fig, axs = plt.subplots(1, 3, figsize=(20, 10))

# Plot the first chart in the first subplot
axs[0].plot(x, y1, label='Line 1', color='blue')
axs[0].set_xlabel('X-Axis')
axs[0].set_ylabel('Y-Axis')
axs[0].set_title('Chart 1')
axs[0].legend()

axs[1].plot(x, y2, label='Line 2', color='green')
axs[1].set_xlabel('X-Axis')
axs[1].set_ylabel('Y-Axis')
axs[1].set_title('Chart 2')
axs[1].legend()

axs[2].plot(x, y3, label='Line 3', color='red')
axs[2].set_xlabel('X-Axis')
axs[2].set_ylabel('Y-Axis')
axs[2].set_title('Chart 3')
axs[2].legend()

plt.show()
```
